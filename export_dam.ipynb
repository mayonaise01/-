```python
import requests
import pandas as pd
import json

from datetime import date, timedelta

def daterange(start_date: date, end_date: date):
    days = int((end_date - start_date).days)
    for n in range(days):
        yield start_date + timedelta(n)

start_date = date(2024, 1, 1)
end_date = date(2024, 10, 1)
```

```python

result = []

headers = { 'Content-Type': 'application/json'  }

for single_date in daterange(start_date, end_date):
  response_data = []

  urls_rsv = f'https://app.rid.go.th/reservoir/api/dam/public/{single_date.strftime("%Y-%m-%d")}'

  response = requests.request("GET", urls_rsv, headers=headers)
  json_data = response.json()  # Parse the JSON response
  response_data.append(json_data)
  for i in response_data[0]['data']:
    for j in i['dam']:
      j['date'] = single_date.strftime("%Y-%m-%d")
      result.append(j)


df = pd.DataFrame(result)

csv_filename = f'dam_{start_date.strftime("%Y")}.csv'
df.to_csv(csv_filename, index=False, encoding="TIS-620")

from google.colab import files
files.download(csv_filename)

print(f"Data written to {csv_filename} and ready for download.")
```
